```python
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
data_base = np.load('data_base_2m.npy',allow_pickle=True).item()
data_fuse = np.load('data_bicross.npy',allow_pickle=True).item()
```

```python
destination = r'G:\EFS-GX6\4130_Arbeitsgruppen\4137_KHO_Performance\Studentische_Themen\RaviKothari\thesis\images'
```

```python
fig,ax = plt.subplots(1,3,figsize =(20,6))
title  =['Pedestrian','Cyclist','Car']

for i in range(3):
    ax[i].plot(data_base[str(i)]['R'],data_base[str(i)]['P'],label='Baseline')
    ax[i].plot(data_fuse[str(i)]['R'],data_fuse[str(i)]['P'],label='Cross Atention')
    ax[i].set_xlim(0,1)
    ax[i].set_ylim(0.6,1.1)
    ax[i].set_xlabel('Recall',fontsize=14)
    ax[i].set_ylabel('Precision',fontsize=14)
    ax[i].title.set_text(title[i])
    ax[i].title.set_size(14)
    ax[i].legend()
plt.savefig(os.path.join(destination,'PvsRcross.jpg'),dpi=1200)
```
